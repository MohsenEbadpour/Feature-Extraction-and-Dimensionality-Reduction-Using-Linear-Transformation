```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("iris.data", header=None)
```

```python
def plot_features(f_1,f_2):
    global Index
    plt.subplot(1,3,Index)
    Index +=1 
    first_class = data.loc[data[4] == "Iris-setosa"]
    second_class = data.loc[data[4] == "Iris-versicolor"]
    third_class = data.loc[data[4] == "Iris-virginica"]
    plt.scatter(first_class[f_1], first_class[f_2],label="Iris-setosa", color = 'hotpink')
    plt.scatter(second_class[f_1], second_class[f_2],label="Iris-versicolor", color = '#88a999')
    plt.scatter(third_class[f_1], third_class[f_2],label="Iris-virginica", color = '#44f888')
    plt.title("Distributions for feature {0} and {1}".format(f_1+1,f_2+1))
    plt.ylabel("Feature {0}".format(f_2+1))
    plt.xlabel("Feature {0}".format(f_1+1))
    plt.legend()
    
Index = 1
plt.figure(figsize=(25, 8))
plot_features(0,3)    
plot_features(1,2)
plot_features(2,3)
plt.show()
```
