```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("iris.data", header=None)
```

```python
Index = 1
plt.figure(figsize=(30, 15))

def plt_without_trans(f_1,f_2):
    global Index
    plt.subplot(3,5,Index)
    Index +=1 
    first_class = data.loc[data[4] == "Iris-setosa"]
    second_class = data.loc[data[4] == "Iris-versicolor"]
    third_class = data.loc[data[4] == "Iris-virginica"]
    plt.scatter(first_class[f_1], first_class[f_2],label="Iris-setosa", color = 'hotpink')
    plt.scatter(second_class[f_1], second_class[f_2],label="Iris-versicolor", color = '#4cbbd6')
    plt.scatter(third_class[f_1], third_class[f_2],label="Iris-virginica", color = '#22e888')
    plt.title("Distributions for feature {0} and {1} without transform".format(f_1+1,f_2+1))
    plt.ylabel("Feature {0}".format(f_2+1))
    plt.xlabel("Feature {0}".format(f_1+1))
    plt.legend()
    
    
def plt_T(f_1,f_2,x_t,y_t):
    global Index
    plt.subplot(3,5,Index)
    Index +=1 
    first_class = data.loc[data[4] == "Iris-setosa"]
    Y = [x*x_t + y*y_t for x,y in zip(first_class[f_1],first_class[f_2])]
    X = [1]*(len(Y))
    #X,Y = Y,X
    plt.hist(Y,label="Iris-setosa", alpha=0.3, color = 'hotpink')
    
    second_class = data.loc[data[4] == "Iris-versicolor"]
    Y = [x*x_t + y*y_t for x,y in zip(second_class[f_1],second_class[f_2])]
    X = [0]*(len(Y))
    #X,Y = Y,X
    plt.hist(Y,label="Iris-versicolor", alpha=0.3, color = '#4cbbd6')
    
    third_class = data.loc[data[4] == "Iris-virginica"]
    Y = [x*x_t + y*y_t for x,y in zip(third_class[f_1],third_class[f_2])]
    X = [-1]*(len(Y))
    #X,Y = Y,X
    plt.hist(Y,label="Iris-virginica", alpha=0.3, color = '#22e888')
    
    plt.title("Distributions for feature {0} and {1} with T({2},{3})".format(f_1+1,f_2+1,x_t,y_t))

    
    plt.legend()



    
    
plt_without_trans(0,1)
plt_T(0,1,1.2,-0.3)
plt_T(0,1,-1.8,0.6)
plt_T(0,1,1.4,0.5)
plt_T(0,1,-0.5,-1)

plt_without_trans(0,2)
plt_T(0,2,1.2,-0.3)
plt_T(0,2,-1.8,0.6)
plt_T(0,2,1.4,0.5)
plt_T(0,2,-0.5,-1)

plt_without_trans(1,3)
plt_T(1,3,1.2,-0.3)
plt_T(1,3,-1.8,0.6)
plt_T(1,3,1.4,0.5)
plt_T(1,3,-0.5,-1)

plt.show()
```
